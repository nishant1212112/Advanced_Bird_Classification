```python
import cv2,os,time,pandas as pd
```

```python
# Used as counter variable 
def FrameCapture(pathVid,ImNum,ImClass): 
    global TrainCount
    global ValidCount
    global TestCount
    global TrainLabels
    global ValidLabels
    global TestLabels

    # Path to video file 
    vidObj = cv2.VideoCapture(pathVid) 
  
    # checks whether frames were extracted 
    success, image = vidObj.read()
    count=0
    while success: 
  
        # vidObj object calls read 
        # function extract frames 
        
        
        # Saves the frames with frame-count 
        if count%100==0:
            if ImNum%5==0:
                #cv2.imwrite("Frames/Test/Image{}.jpg".format(TestCount), image)
                TestLabels.append(ImClass)
                TestCount+=1
            elif ImNum%5==1:
                #cv2.imwrite("Frames/Valid/Image{}.jpg".format(ValidCount), image)
                ValidLabels.append(ImClass)
                ValidCount+=1
            else:
                #cv2.imwrite("Frames/Train/Image{}.jpg".format(TrainCount), image)
                TrainLabels.append(ImClass)
                TrainCount+=1

        count += 1
        success, image = vidObj.read() 
        
    return count/100
```

```python
TrainCount = 0
TestCount = 0
ValidCount= 0
TrainLabels=[]
ValidLabels=[]
TestLabels=[]
for index in range(1,23):
    VidFolPath= "VB 100/vb100_video{}".format(index)
    for folder in os.listdir(VidFolPath):
        if folder != '.DS_Store':
            print folder
            start=time.time()
            print "start:",0.00
            for i,Videos in enumerate(os.listdir(VidFolPath+'/'+folder)):
                print FrameCapture(VidFolPath+'/'+folder+'/'+Videos,i,folder)," Frames extracted"
                print i+1,":",time.time()-start
```
